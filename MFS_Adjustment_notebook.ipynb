```python
import cx_Oracle
conn = cx_Oracle.connect('aws_reader/mfaws2@10.68.72.101:1521/mfaws2.epnet.com')
print (conn.version)
cur = conn.cursor()


ora_list = []
print('start loading oracle data')
#cur.execute('select an from article_sch order by an')
sql_query = '''Select ip.MID,ip.DTFORMAT,i.CUID,ip.PROCNUM,ip.dtmagstart,ip.dtdataend,i.dtprodavail,i.dtdeleted,ip.lost,ip.issueproccomment
From Issueproc ip
Join issue i on i.mid = ip.mid AND i.dtformat = ip.dtformat
WHERE ip.procnum = 13
AND ip.DTDATAEND is null
AND ip.DTMAGSTART is not null
and  (i.dtprodavail is null or i.dtprodavail <= CURRENT_DATE)'''
#cur.execute('Select ip.MID, ip.DTFORMAT, i.CUID,ip.PROCNUM, ip.dtmagstart, ip.dtdataend, i.dtprodavail, i.dtdeleted, ip.lost, ip.issueproccomment From Issueproc ip Join issue i on i.mid = ip.mid AND i.dtformat = ip.dtformat WHERE ip.procnum = 13 AND ip.DTDATAEND is null AND ip.DTMAGSTART is not null')
#print(sql_query)
cur.execute(sql_query)

for result in cur.fetchall():
    res_list = list ((str(result[0]),str(result[1]),str(result[2]),str(result[3]),str(result[4]),str(result[5]),str(result[6]),str(result[7]),str(result[8]),str(result[9])))
    ora_list.append(res_list)
#print (list)
cur.close
conn.close()
print('finish loading oracle')
print('length of list: ', len(ora_list))
```
